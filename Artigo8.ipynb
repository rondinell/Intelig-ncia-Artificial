```python
# - Rondinelli Alves de Andrade -*-
"""
Programa Colab: Redes Neurais para Previsão Financeira (LSTM)

Este script demonstra como usar uma Rede Neural Recorrente (LSTM)
para prever o preço de fechamento de um ativo financeiro,
abordando o desafio da complexidade e não-linearidade das séries temporais.
"""

# 1. Instalação de Bibliotecas Necessárias
# Usamos --quiet para não poluir a saída do Colab com mensagens de instalação.
!pip install yfinance --quiet
!pip install scikit-learn --quiet
!pip install tensorflow --quiet
!pip install matplotlib --quiet
!pip install numpy --quiet
!pip install pandas --quiet

# 2. Importação das Bibliotecas
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import datetime

print("Bibliotecas importadas com sucesso!")

# --- Definição do Ativo e Período de Análise ---
TICKER = "BBDC4.SA" # Exemplo: Bradesco. Sinta-se à vontade para mudar!
START_DATE = "2020-01-01"
END_DATE = datetime.datetime.now().strftime('%Y-%m-%d') # Data atual

print(f"\nAnalisando o ativo: {TICKER} de {START_DATE} até {END_DATE}")

# --- 3. O Problema: Complexidade das Séries Temporais Financeiras ---
print("\n--- O Desafio no Mercado Financeiro: Previsão de Séries Temporais ---")
print("Os preços dos ativos financeiros são notoriamente difíceis de prever devido a:")
print("  - Não-linearidade: Não seguem padrões simples e lineares.")
print("  - Não-estacionariedade: Suas propriedades estatísticas mudam ao longo do tempo.")
print("  - Volatilidade: Flutuações rápidas e imprevisíveis.")
print("  - Eventos externos: Notícias, crises, decisões políticas impactam os preços de forma abrupta.")
print("Modelos estatísticos tradicionais muitas vezes lutam para capturar essa complexidade.")

# --- 4. Coleta de Dados Históricos ---
try:
    df = yf.download(TICKER, start=START_DATE, end=END_DATE)
    if df.empty:
        raise ValueError("Nenhum dado baixado. Verifique o ticker ou o período.")
    print(f"\nDados de {TICKER} baixados com sucesso. Total de {len(df)} linhas.")
    # Usaremos apenas o preço de fechamento para esta demonstração
    data = df['Close'].values.reshape(-1, 1)
except Exception as e:
    print(f"Erro ao baixar dados do Yahoo Finance: {e}")
    print("Certifique-se de que o ticker é válido e a conexão com a internet está funcionando.")
    exit() # Encerrar o script se não houver dados

# Visualização inicial da série temporal
plt.figure(figsize=(14, 7))
plt.plot(df.index, data, label='Preço de Fechamento Histórico')
plt.title(f'Série Temporal do Preço de Fechamento de {TICKER}')
plt.xlabel('Data')
plt.ylabel('Preço (R$)')
plt.grid(True)
plt.legend()
plt.show()

# --- 5. Pré-processamento dos Dados: Preparando para a Rede Neural ---
# Normalização dos dados (MinMaxScaler é essencial para redes neurais)
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

# Divisão dos dados em conjuntos de treino e teste
# Usaremos 80% para treino e 20% para teste
train_size = int(len(scaled_data) * 0.8)
train_data = scaled_data[0:train_size, :]
test_data = scaled_data[train_size:len(scaled_data), :]

# Função para criar sequências de entrada e saída (X e Y) para LSTM
# Uma sequência (look_back) de preços passados para prever o próximo preço.
def create_sequences(dataset, look_back=1):
    X, Y = [], []
    for i in range(len(dataset) - look_back):
        a = dataset[i:(i + look_back), 0]
        X.append(a)
        Y.append(dataset[i + look_back, 0])
    return np.array(X), np.array(Y)

LOOK_BACK = 90 # Número de dias anteriores para prever o próximo dia

X_train, y_train = create_sequences(train_data, LOOK_BACK)
X_test, y_test = create_sequences(test_data, LOOK_BACK)

# Reshape dos dados para o formato esperado pela LSTM: [amostras, timesteps, features]
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

print(f"\nDados pré-processados para treinamento da LSTM:")
print(f"  - X_train shape: {X_train.shape}")
print(f"  - y_train shape: {y_train.shape}")
print(f"  - X_test shape: {X_test.shape}")
print(f"  - y_test shape: {y_test.shape}")

# --- 6. A Solução: Construindo a Rede Neural LSTM ---
print("\n--- Construindo o Modelo LSTM: A Solução de Rede Neural ---")
print("A Rede Neural Recorrente (LSTM) é escolhida por sua capacidade de:")
print("  - Lembrar informações de longo prazo em sequências de dados.")
print("  - Capturar padrões temporais complexos e não-lineares.")
print("  - Lidar com o problema de 'gradiente evanescente' comum em RNNs simples.")

model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(LOOK_BACK, 1)))
model.add(Dropout(0.2)) # Previne overfitting
model.add(LSTM(units=50, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(units=1)) # Camada de saída para a previsão do próximo preço

model.compile(optimizer='adam', loss='mean_squared_error')
model.summary()

# --- 7. Treinamento do Modelo ---
print("\n--- Treinando o Modelo LSTM ---")
# Callbacks para melhorar o treinamento:
# EarlyStopping: Para parar o treinamento se o erro não diminuir (evita overfitting)
# ReduceLROnPlateau: Para reduzir a taxa de aprendizado se o erro estagnar
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=0.00001)

history = model.fit(X_train, y_train,
                    epochs=100, # Número máximo de épocas
                    batch_size=32,
                    validation_split=0.1, # Usar 10% do treino para validação
                    callbacks=[early_stopping, reduce_lr],
                    verbose=1)

# Visualizar o histórico de treinamento
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Loss de Treinamento')
plt.plot(history.history['val_loss'], label='Loss de Validação')
plt.title('Histórico de Treinamento do Modelo LSTM')
plt.xlabel('Época')
plt.ylabel('Loss (Erro Quadrático Médio)')
plt.legend()
plt.grid(True)
plt.show()

# --- 8. Realizando Previsões e Avaliando o Modelo ---
print("\n--- Realizando Previsões e Avaliando o Modelo ---")

# Fazendo previsões nos dados de teste
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)

# Invertendo a normalização para obter os valores de preço reais
train_predict = scaler.inverse_transform(train_predict)
test_predict = scaler.inverse_transform(test_predict)
y_train_actual = scaler.inverse_transform(y_train.reshape(-1, 1))
y_test_actual = scaler.inverse_transform(y_test.reshape(-1, 1))

# Calculando o Erro Quadrático Médio (RMSE)
rmse_train = np.sqrt(mean_squared_error(y_train_actual, train_predict))
rmse_test = np.sqrt(mean_squared_error(y_test_actual, test_predict))

print(f"RMSE (Root Mean Squared Error) no conjunto de Treinamento: {rmse_train:.2f}")
print(f"RMSE (Root Mean Squared Error) no conjunto de Teste: {rmse_test:.2f}")

# --- 9. Visualização dos Resultados ---
print("\n--- Visualização das Previsões vs. Valores Reais ---")

# Criando um array para plotar todas as previsões
look_back_offset = LOOK_BACK # Usado para ajustar o índice de plotagem devido ao look_back

# Criar arrays para armazenar as previsões alinhadas com o DataFrame original
train_plot = np.empty_like(data)
train_plot[:, :] = np.nan
train_plot[look_back_offset:len(train_predict) + look_back_offset, :] = train_predict

test_plot = np.empty_like(data)
test_plot[:, :] = np.nan
test_plot[len(train_predict) + (look_back_offset * 2) - 1:len(data) - 1, :] = test_predict

# Plotando os resultados
plt.figure(figsize=(16, 8))
plt.plot(df.index, scaler.inverse_transform(scaled_data), label='Preço Real', color='blue', alpha=0.7)
plt.plot(df.index[look_back_offset : len(train_predict) + look_back_offset], train_predict, label='Previsão Treino', color='green', linestyle='--')
plt.plot(df.index[len(train_predict) + (look_back_offset * 2) - 1 : len(data) - 1], test_predict, label='Previsão Teste', color='red', linestyle='--')
plt.title(f'Previsão de Preço de Fechamento de {TICKER} com LSTM')
plt.xlabel('Data')
plt.ylabel('Preço (R$)')
plt.legend()
plt.grid(True)
plt.show()

print("\n--- Conclusão da Demonstração ---")
print("Este programa demonstrou como uma Rede Neural LSTM pode ser aplicada para prever séries temporais financeiras.")
print("Enquanto a previsão exata de preços é um desafio inerentemente complexo e incerto no mercado financeiro,")
print("modelos como as LSTMs podem capturar padrões complexos e fornecer insights valiosos,")
print("superando as limitações de abordagens lineares e contribuindo para sistemas de tomada de decisão mais sofisticados.")
print("Lembre-se que este é um modelo simplificado e não deve ser usado para decisões de investimento reais sem validação e análise aprofundadas.")

```
