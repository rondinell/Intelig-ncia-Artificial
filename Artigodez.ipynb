```python
# ============================================================
# LSTM para Predição de Ações B3 - Google Colab
# ============================================================

# Instalar bibliotecas
!pip install yfinance ta scikit-learn --quiet

# Importar bibliotecas básicas
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Importar bibliotecas de ML
try:
    import ta
    from sklearn.preprocessing import MinMaxScaler
    from sklearn.metrics import classification_report, confusion_matrix
    print("✓ Bibliotecas básicas carregadas")
except ImportError as e:
    print(f"Erro ao importar: {e}")

# Importar TensorFlow
try:
    import tensorflow as tf
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import LSTM, Dense, Dropout
    from tensorflow.keras.callbacks import EarlyStopping
    print(f"✓ TensorFlow {tf.__version__} carregado")
except ImportError as e:
    print(f"Erro ao importar TensorFlow: {e}")

# Configurações
np.random.seed(42)
tf.random.set_seed(42)

# ============================================================
# 1. BAIXAR DADOS
# ============================================================
print("1. Baixando dados...")
ticker = 'VALE3.SA'

try:
    data = yf.download(ticker, start='2020-01-01', end='2024-01-01', progress=False)
    print(f"✓ {len(data)} registros baixados para {ticker}")
except Exception as e:
    print(f"Erro ao baixar dados: {e}")
    # Usar dados de exemplo se falhar
    dates = pd.date_range('2020-01-01', '2024-01-01', freq='D')
    data = pd.DataFrame({
        'Open': np.random.uniform(20, 30, len(dates)),
        'High': np.random.uniform(25, 35, len(dates)),
        'Low': np.random.uniform(15, 25, len(dates)),
        'Close': np.random.uniform(20, 30, len(dates)),
        'Volume': np.random.uniform(1000000, 5000000, len(dates))
    }, index=dates)

# ============================================================
# 2. PREPARAR DADOS
# ============================================================
print("2. Preparando dados...")

# Usar apenas preço de fechamento
prices = data['Close'].values.reshape(-1, 1)

# Normalizar dados
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(prices)

# Criar sequências
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(seq_length, len(data)):
        X.append(data[i-seq_length:i, 0])
        y.append(1 if data[i, 0] > data[i-1, 0] else 0)  # 1 se subiu, 0 se desceu
    return np.array(X), np.array(y)

SEQ_LENGTH = 60
X, y = create_sequences(scaled_data, SEQ_LENGTH)

print(f"✓ Sequências criadas: X shape = {X.shape}, y shape = {y.shape}")

# Dividir dados
train_size = int(len(X) * 0.8)
X_train = X[:train_size]
y_train = y[:train_size]
X_test = X[train_size:]
y_test = y[train_size:]

# Reshape para LSTM
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

print(f"✓ Dados divididos - Treino: {X_train.shape[0]}, Teste: {X_test.shape[0]}")

# ============================================================
# 3. CRIAR MODELO
# ============================================================
print("3. Criando modelo...")

model = Sequential([
    LSTM(100, return_sequences=True, input_shape=(SEQ_LENGTH, 1)),
    Dropout(0.25),
    LSTM(100, return_sequences=False),
    Dropout(0.25),
    Dense(25),
    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

print("✓ Modelo criado")
model.summary()

# ============================================================
# 4. TREINAR MODELO
# ============================================================
print("4. Treinando modelo...")

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)

try:
    history = model.fit(
        X_train, y_train,
        batch_size=32,
        epochs=50,
        validation_data=(X_test, y_test),
        callbacks=[early_stop],
        verbose=1
    )
    print("✓ Treinamento concluído")
except Exception as e:
    print(f"Erro no treinamento: {e}")

# ============================================================
# 5. AVALIAR MODELO
# ============================================================
print("5. Avaliando modelo...")

# Fazer predições
predictions = model.predict(X_test)
y_pred = (predictions > 0.5).astype(int).flatten()

# Calcular métricas
accuracy = np.mean(y_pred == y_test)
print(f"✓ Acurácia: {accuracy:.3f}")

# Relatório detalhado
print("\nRelatório de Classificação:")
print(classification_report(y_test, y_pred, target_names=['Desceu', 'Subiu']))

print("\nMatriz de Confusão:")
cm = confusion_matrix(y_test, y_pred)
print(cm)

# ============================================================
# 6. VISUALIZAR RESULTADOS
# ============================================================
print("6. Criando visualizações...")

# Criar gráficos
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Gráfico 1: Histórico de treinamento - Acurácia
axes[0,0].plot(history.history['accuracy'], label='Treino', color='blue')
axes[0,0].plot(history.history['val_accuracy'], label='Validação', color='red')
axes[0,0].set_title('Acurácia do Modelo')
axes[0,0].set_xlabel('Épocas')
axes[0,0].set_ylabel('Acurácia')
axes[0,0].legend()
axes[0,0].grid(True)

# Gráfico 2: Histórico de treinamento - Loss
axes[0,1].plot(history.history['loss'], label='Treino', color='blue')
axes[0,1].plot(history.history['val_loss'], label='Validação', color='red')
axes[0,1].set_title('Loss do Modelo')
axes[0,1].set_xlabel('Épocas')
axes[0,1].set_ylabel('Loss')
axes[0,1].legend()
axes[0,1].grid(True)

# Gráfico 3: Matriz de Confusão
im = axes[1,0].imshow(cm, interpolation='nearest', cmap='Blues')
axes[1,0].set_title('Matriz de Confusão')
tick_marks = np.arange(2)
axes[1,0].set_xticks(tick_marks)
axes[1,0].set_xticklabels(['Desceu', 'Subiu'])
axes[1,0].set_yticks(tick_marks)
axes[1,0].set_yticklabels(['Desceu', 'Subiu'])

# Adicionar valores na matriz
for i in range(2):
    for j in range(2):
        axes[1,0].text(j, i, str(cm[i, j]),
                      ha='center', va='center', color='white' if cm[i, j] > cm.max()/2 else 'black')

axes[1,0].set_xlabel('Predição')
axes[1,0].set_ylabel('Real')

# Gráfico 4: Preços históricos
recent_prices = data['Close'][-200:].values
axes[1,1].plot(recent_prices, color='blue', linewidth=1)
axes[1,1].set_title('Preços Históricos Recentes')
axes[1,1].set_xlabel('Dias')
axes[1,1].set_ylabel('Preço (R$)')
axes[1,1].grid(True)

plt.tight_layout()
plt.show()

# ============================================================
# 7. ESTATÍSTICAS FINAIS
# ============================================================
print("7. Estatísticas finais:")
print(f"Total de amostras testadas: {len(y_test)}")
print(f"Predições corretas: {np.sum(y_pred == y_test)}")
print(f"Predições incorretas: {np.sum(y_pred != y_test)}")
print(f"Acurácia final: {accuracy:.1%}")

# Distribuição das classes
print(f"Distribuição real - Desceu: {np.sum(y_test == 0)}, Subiu: {np.sum(y_test == 1)}")
print(f"Distribuição predita - Desceu: {np.sum(y_pred == 0)}, Subiu: {np.sum(y_pred == 1)}")

# ============================================================
# 8. PREDIÇÃO PARA PRÓXIMO DIA
# ============================================================
print("\n8. Fazendo predição para o próximo dia...")

# Pegar os últimos 60 dias para fazer a predição
last_sequence = scaled_data[-SEQ_LENGTH:].reshape(1, SEQ_LENGTH, 1)

# Fazer predição
next_day_prob = model.predict(last_sequence, verbose=0)[0][0]
next_day_prediction = 1 if next_day_prob > 0.5 else 0

# Mostrar resultado
print("="*60)
print("🔮 PREDIÇÃO PARA O PRÓXIMO DIA DE PREGÃO")
print("="*60)
print(f"📈 Ação analisada: {ticker}")
print(f"💰 Último preço: R$ {data['Close'].iloc[-1].values[0]:.2f}") # Access the value from the Series
print(f"🎯 Probabilidade de SUBIR: {next_day_prob:.1%}")
print(f"🎯 Probabilidade de DESCER: {(1-next_day_prob):.1%}")
print()

if next_day_prediction == 1:
    print("🟢 PREDIÇÃO: A AÇÃO VAI SUBIR! 📈")
    print("💡 Sugestão: COMPRAR")
    if next_day_prob > 0.7:
        print("🔥 Confiança: ALTA")
    elif next_day_prob > 0.6:
        print("⚡ Confiança: MÉDIA")
    else:
        print("⚠️  Confiança: BAIXA")
else:
    print("🔴 PREDIÇÃO: A AÇÃO VAI DESCER! 📉")
    print("💡 Sugestão: VENDER ou AGUARDAR")
    if next_day_prob < 0.3:
        print("🔥 Confiança: ALTA")
    elif next_day_prob < 0.4:
        print("⚡ Confiança: MÉDIA")
    else:
        print("⚠️  Confiança: BAIXA")

print("="*60)

# ============================================================
# 9. ANÁLISE DE TENDÊNCIA (últimos 5 dias)
# ============================================================
print("\n9. Análise de tendência recente...")

# Fazer predições para os últimos 5 dias (para verificar padrão)
recent_predictions = []
for i in range(5, 0, -1):
    seq = scaled_data[-SEQ_LENGTH-i:-i].reshape(1, SEQ_LENGTH, 1)
    prob = model.predict(seq, verbose=0)[0][0]
    recent_predictions.append(prob)

print("\n📊 TENDÊNCIA DOS ÚLTIMOS 5 DIAS:")
print("-" * 40)
for i, prob in enumerate(recent_predictions):
    day = f"Dia -{5-i}"
    direction = "📈 SUBIR" if prob > 0.5 else "📉 DESCER"
    confidence = "🔥" if abs(prob - 0.5) > 0.2 else "⚡" if abs(prob - 0.5) > 0.1 else "⚠️"
    print(f"{day}: {direction} ({prob:.1%}) {confidence}")

# Calcular tendência geral
avg_prob = np.mean(recent_predictions)
trend = "ALTA" if avg_prob > 0.55 else "BAIXA" if avg_prob < 0.45 else "LATERAL"

print(f"\n🎯 TENDÊNCIA GERAL: {trend}")
print(f"📊 Probabilidade média de alta: {avg_prob:.1%}")

# ============================================================
# 10. AVISO IMPORTANTE
# ============================================================
print("\n" + "="*60)
print("⚠️  AVISO IMPORTANTE")
print("="*60)
print("🚨 Esta é apenas uma predição baseada em machine learning")
print("📊 A acurácia do modelo foi de {:.1%}".format(accuracy))
print("💡 NÃO use como única base para decisões de investimento")
print("🔍 Sempre faça sua própria análise fundamentalista")
print("💰 Invista apenas o que pode perder")
print("📈 Mercado de ações envolve riscos")
print("="*60)

print("\n🎉 Análise completa concluída!")
```
