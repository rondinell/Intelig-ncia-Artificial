```python
# ====================================================================
# Parte 1: INSTALAÇÃO E IMPORTAÇÕES
# ====================================================================
print("Instalando bibliotecas necessárias...")
!pip install yfinance ta --quiet

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ta
import os
import shutil
from datetime import date
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Rescaling
from tensorflow.keras.utils import image_dataset_from_directory, load_img, img_to_array
from tensorflow.keras.callbacks import EarlyStopping

print("Bibliotecas carregadas com sucesso!")

# ====================================================================
# Parte 2: CONFIGURAÇÕES GLOBAIS
# ====================================================================
print("\nDefinindo configurações globais...")
# --- Configurações do Ativo e Datas ---
TICKER = 'VALE3.SA'
START_DATE = '2015-01-01'
END_DATE = date.today().strftime("%Y-%m-%d")

# --- Configurações para Geração das Imagens e Rótulos ---
SEQ_LENGTH = 60
FUTURE_DAYS = 5
THRESHOLD = 0.02

# --- Configurações das Imagens e do Modelo ---
IMG_SIZE = 128
BATCH_SIZE = 32
EPOCHS = 30 # Aumentamos as épocas, pois o Early Stopping vai parar quando for o ideal

# --- Configurações de Pastas ---
BASE_DIR = '/content/stock_data'
if os.path.exists(BASE_DIR):
    shutil.rmtree(BASE_DIR)

os.makedirs(os.path.join(BASE_DIR, 'train', 'ALTA'), exist_ok=True)
os.makedirs(os.path.join(BASE_DIR, 'train', 'BAIXA'), exist_ok=True)
os.makedirs(os.path.join(BASE_DIR, 'train', 'NEUTRA'), exist_ok=True)

print("Configurações definidas e pastas criadas.")

# ====================================================================
# Parte 3: DOWNLOAD E PREPARAÇÃO DOS DADOS
# ====================================================================
print(f"\nBaixando dados históricos para {TICKER}...")
data = yf.download(TICKER, start=START_DATE, end=END_DATE, progress=False)
data.dropna(inplace=True)
try:
    # Tenta remover o MultiIndex se ele existir
    data.columns = data.columns.get_level_values(0)
except AttributeError:
    # Se não for MultiIndex, não faz nada
    print("Nomes de colunas já estão no formato correto.")

# Adiciona todos os indicadores técnicos da biblioteca 'ta'
print("Calculando indicadores técnicos...")
data = ta.add_all_ta_features(data, open="Open", high="High", low="Low", close="Close", volume="Volume", fillna=True)
print(f"{len(data)} dias de dados baixados e processados.")


# ====================================================================
# Parte 4: GERADOR DE IMAGENS COM MÚLTIPLOS GRÁFICOS
# ====================================================================
print("\nIniciando geração de imagens e rótulos (com múltiplos gráficos)...")
count = {'ALTA': 0, 'BAIXA': 0, 'NEUTRA': 0}

for i in range(len(data) - SEQ_LENGTH - FUTURE_DAYS):
    dados_janela = data.iloc[i : i + SEQ_LENGTH]

    preco_final_janela = dados_janela.iloc[-1]['Close']
    preco_futuro = data.iloc[i + SEQ_LENGTH + FUTURE_DAYS - 1]['Close']

    label = 'NEUTRA'
    if preco_futuro > preco_final_janela * (1 + THRESHOLD):
        label = 'ALTA'
    elif preco_futuro < preco_final_janela * (1 - THRESHOLD):
        label = 'BAIXA'
    count[label] += 1

    pasta_destino = os.path.join(BASE_DIR, 'train', label)
    nome_arquivo = f"{data.index[i].strftime('%Y-%m-%d')}.png"
    caminho_completo = os.path.join(pasta_destino, nome_arquivo)

    # Criação de 3 subplots empilhados verticalmente
    fig, ax = plt.subplots(
        3, 1,
        figsize=(IMG_SIZE/100, IMG_SIZE/100),
        dpi=100,
        gridspec_kw={'height_ratios': [2, 1, 1]} # Proporção: Preço (maior), RSI (menor), Volume (menor)
    )

    # --- PLOT 1: Preço e Bandas de Bollinger ---
    ax[0].plot(dados_janela['Close'].values, color='black', linewidth=1.0)
    ax[0].plot(dados_janela['volatility_bbh'].values, color='blue', linewidth=0.5, alpha=0.5)
    ax[0].plot(dados_janela['volatility_bbl'].values, color='blue', linewidth=0.5, alpha=0.5)
    ax[0].plot(dados_janela['volatility_bbm'].values, color='red', linewidth=0.5, alpha=0.7, linestyle='--')

    # --- PLOT 2: RSI (Índice de Força Relativa) ---
    ax[1].plot(dados_janela['momentum_rsi'].values, color='purple', linewidth=0.8)
    ax[1].axhline(70, color='gray', linestyle='--', linewidth=0.5, alpha=0.8)
    ax[1].axhline(30, color='gray', linestyle='--', linewidth=0.5, alpha=0.8)

    # --- PLOT 3: Volume ---
    ax[2].bar(range(len(dados_janela)), dados_janela['Volume'].values, color='gray', alpha=0.6)

    # Remove os eixos de todos os gráficos
    ax[0].axis('off')
    ax[1].axis('off')
    ax[2].axis('off')

    # Ajusta o espaçamento para remover bordas brancas
    plt.subplots_adjust(left=0, right=1, top=1, bottom=0, hspace=0.05)

    fig.savefig(caminho_completo, bbox_inches='tight', pad_inches=0)
    plt.close(fig)

print("Geração de imagens aprimoradas concluída!")
print(f"Total de imagens geradas: {sum(count.values())}")
print(f"Distribuição: {count}")

# ====================================================================
# Parte 5: PREPARAÇÃO DOS DADOS PARA O MODELO
# ====================================================================
print("\nCarregando e preparando os datasets de imagens...")

train_dataset_raw = image_dataset_from_directory(
    os.path.join(BASE_DIR, 'train'),
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE
)
class_names = train_dataset_raw.class_names

validation_dataset_raw = image_dataset_from_directory(
    os.path.join(BASE_DIR, 'train'),
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE
)

AUTOTUNE = tf.data.AUTOTUNE
train_dataset = train_dataset_raw.cache().prefetch(buffer_size=AUTOTUNE)
validation_dataset = validation_dataset_raw.cache().prefetch(buffer_size=AUTOTUNE)

print(f"Classes encontradas: {class_names}")

# ====================================================================
# Parte 6: CONSTRUÇÃO DO MODELO CNN
# ====================================================================
print("\nConstruindo o modelo CNN...")

model_cnn = Sequential([
    Input(shape=(IMG_SIZE, IMG_SIZE, 3)),
    Rescaling(1./255),
    Conv2D(32, 3, padding='same', activation='relu'),
    MaxPooling2D(),
    Conv2D(64, 3, padding='same', activation='relu'),
    MaxPooling2D(),
    Conv2D(128, 3, padding='same', activation='relu'),
    MaxPooling2D(),
    Dropout(0.5),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(len(class_names), activation='softmax')
])

model_cnn.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model_cnn.summary()

# ====================================================================
# Parte 7: TREINAMENTO DO MODELO
# ====================================================================
print("\nIniciando o treinamento da CNN...")

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

history = model_cnn.fit(
    train_dataset,
    validation_data=validation_dataset,
    epochs=EPOCHS,
    callbacks=[early_stopping]
)
print("Treinamento concluído!")

# ====================================================================
# Parte 8: AVALIAÇÃO E PREVISÃO
# ====================================================================
print("\nGerando gráficos de avaliação...")

acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

# Define o range de épocas baseado no que realmente rodou
epochs_range = range(len(acc))

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Acurácia de Treino')
plt.plot(epochs_range, val_acc, label='Acurácia de Validação')
plt.legend(loc='lower right')
plt.title('Acurácia de Treino e Validação')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Perda de Treino')
plt.plot(epochs_range, val_loss, label='Perda de Validação')
plt.legend(loc='upper right')
plt.title('Perda de Treino e Validação')
plt.show()

print("\nFazendo previsão com os dados mais recentes...")
dados_recentes = data.iloc[-SEQ_LENGTH:]

fig, ax = plt.subplots(3, 1, figsize=(IMG_SIZE/100, IMG_SIZE/100), dpi=100, gridspec_kw={'height_ratios': [2, 1, 1]})
ax[0].plot(dados_recentes['Close'].values, color='black', linewidth=1.0)
ax[0].plot(dados_recentes['volatility_bbh'].values, color='blue', linewidth=0.5, alpha=0.5)
ax[0].plot(dados_recentes['volatility_bbl'].values, color='blue', linewidth=0.5, alpha=0.5)
ax[0].plot(dados_recentes['volatility_bbm'].values, color='red', linewidth=0.5, alpha=0.7, linestyle='--')
ax[1].plot(dados_recentes['momentum_rsi'].values, color='purple', linewidth=0.8)
ax[1].axhline(70, color='gray', linestyle='--', linewidth=0.5, alpha=0.8)
ax[1].axhline(30, color='gray', linestyle='--', linewidth=0.5, alpha=0.8)
ax[2].bar(range(len(dados_recentes)), dados_recentes['Volume'].values, color='gray', alpha=0.6)
ax[0].axis('off')
ax[1].axis('off')
ax[2].axis('off')
plt.subplots_adjust(left=0, right=1, top=1, bottom=0, hspace=0.05)

caminho_predicao = '/content/pred_temp.png'
fig.savefig(caminho_predicao, bbox_inches='tight', pad_inches=0)
plt.close(fig)

img = load_img(caminho_predicao, target_size=(IMG_SIZE, IMG_SIZE))
img_array = img_to_array(img)
img_array = tf.expand_dims(img_array, 0)

predictions = model_cnn.predict(img_array)
score = tf.nn.softmax(predictions[0])

print("\n" + "="*60)
print(f"🔮 PREDIÇÃO PARA OS PRÓXIMOS {FUTURE_DAYS} DIAS DE {TICKER}")
print("="*60)
for i in range(len(class_names)):
    print(f"Probabilidade de {class_names[i]}: {100 * score[i]:.2f}%")
predicted_class = class_names[np.argmax(score)]
print(f"\n🟢 TENDÊNCIA MAIS PROVÁVEL: {predicted_class}")
print("="*60)

# ====================================================================
# Parte 9: PREPARAÇÃO DOS DADOS PARA O MODELO LSTM
# ====================================================================
print("\nPreparando dados para o modelo LSTM...")
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

# Selecionar as features (colunas) que o modelo LSTM usará
# Vamos usar as mesmas colunas usadas para gerar os gráficos e mais algumas
features = ['Close', 'Volume', 'volatility_bbh', 'volatility_bbl', 'volatility_bbm', 'momentum_rsi']
data_lstm = data[features].copy()

# Normalizar os dados para que fiquem entre 0 and 1
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data_lstm)

# --- Criar as sequências e os rótulos (labels) ---
X_lstm, y_lstm = [], []
for i in range(len(scaled_data) - SEQ_LENGTH - FUTURE_DAYS):
    # Sequência de dados (janela de SEQ_LENGTH dias)
    X_lstm.append(scaled_data[i : i + SEQ_LENGTH])

    # Rótulo (label) - baseado na variação do preço futuro
    preco_final_janela = data['Close'].iloc[i + SEQ_LENGTH -1]
    preco_futuro = data['Close'].iloc[i + SEQ_LENGTH + FUTURE_DAYS - 1]

    label_val = 2 # NEUTRA
    if preco_futuro > preco_final_janela * (1 + THRESHOLD):
        label_val = 0 # ALTA
    elif preco_futuro < preco_final_janela * (1 - THRESHOLD):
        label_val = 1 # BAIXA
    y_lstm.append(label_val)

X_lstm, y_lstm = np.array(X_lstm), np.array(y_lstm)

# Dividir em dados de treino e validação
X_train_lstm, X_val_lstm, y_train_lstm, y_val_lstm = train_test_split(
    X_lstm, y_lstm, test_size=0.2, random_state=123, stratify=y_lstm
)

print(f"Formato dos dados de treino (X): {X_train_lstm.shape}")
print(f"Formato dos dados de treino (y): {y_train_lstm.shape}")
print("Dados para LSTM prontos!")

# ====================================================================
# Parte 10: CONSTRUÇÃO E TREINAMENTO DO MODELO LSTM
# ====================================================================
print("\nConstruindo e treinando o modelo LSTM...")
from tensorflow.keras.layers import LSTM

model_lstm = Sequential([
    Input(shape=(X_train_lstm.shape[1], X_train_lstm.shape[2])), # (SEQ_LENGTH, num_features)
    LSTM(units=50, return_sequences=True),
    Dropout(0.2),
    LSTM(units=50, return_sequences=False),
    Dropout(0.2),
    Dense(units=25, activation='relu'),
    Dense(len(class_names), activation='softmax') # Saída com 3 classes (ALTA, BAIXA, NEUTRA)
])

model_lstm.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model_lstm.summary()

# Treinamento do modelo LSTM
history_lstm = model_lstm.fit(
    X_train_lstm, y_train_lstm,
    validation_data=(X_val_lstm, y_val_lstm),
    epochs=EPOCHS, # Usando as mesmas épocas e Early Stopping
    batch_size=BATCH_SIZE,
    callbacks=[early_stopping]
)
print("Treinamento do LSTM concluído!")

# ====================================================================
# Parte 11: AVALIAÇÃO E PREVISÃO COM O MODELO LSTM
# ====================================================================
print("\nAvaliando o modelo LSTM...")

# Gráficos de Acurácia e Perda para o LSTM
acc_lstm = history_lstm.history['accuracy']
val_acc_lstm = history_lstm.history['val_accuracy']
loss_lstm = history_lstm.history['loss']
val_loss_lstm = history_lstm.history['val_loss']

epochs_range_lstm = range(len(acc_lstm))

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs_range_lstm, acc_lstm, label='Acurácia de Treino (LSTM)')
plt.plot(epochs_range_lstm, val_acc_lstm, label='Acurácia de Validação (LSTM)')
plt.legend(loc='lower right')
plt.title('Acurácia do Modelo LSTM')

plt.subplot(1, 2, 2)
plt.plot(epochs_range_lstm, loss_lstm, label='Perda de Treino (LSTM)')
plt.plot(epochs_range_lstm, val_loss_lstm, label='Perda de Validação (LSTM)')
plt.legend(loc='upper right')
plt.title('Perda do Modelo LSTM')
plt.show()


# --- Fazer previsão com os dados mais recentes ---
print("\nFazendo previsão com o modelo LSTM...")
dados_recentes_lstm = data_lstm.iloc[-SEQ_LENGTH:]
scaled_dados_recentes = scaler.transform(dados_recentes_lstm)
X_pred_lstm = np.expand_dims(scaled_dados_recentes, axis=0) # Adiciona a dimensão do batch

predictions_lstm = model_lstm.predict(X_pred_lstm)
score_lstm = tf.nn.softmax(predictions_lstm[0])

print("\n" + "="*60)
print(f"🔮 PREDIÇÃO (LSTM) PARA OS PRÓXIMOS {FUTURE_DAYS} DIAS DE {TICKER}")
print("="*60)
for i in range(len(class_names)):
    print(f"Probabilidade de {class_names[i]}: {100 * score_lstm[i]:.2f}%")
predicted_class_lstm = class_names[np.argmax(score_lstm)]
print(f"\n🟢 TENDÊNCIA MAIS PROVÁVEL (LSTM): {predicted_class_lstm}")
print("="*60)
```
